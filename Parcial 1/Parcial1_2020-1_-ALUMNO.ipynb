## Ej 3: Matrices

Cree las matrices:  

$M=\begin{bmatrix} 2 & 1 & 4 \\ -3 & 0 & 5  \end{bmatrix}$  
$N=\begin{bmatrix} -2 & 1 \\ 0 & 3 \\ 1 & 4    \end{bmatrix}$ 

```python
# Escriba aquí su código
M=np.array([[2,1,4],[-3,0,5]])
N=np.array([[-2,1],[0,3],[1,4] ])
M,N
```

Calcule: 

$ Q1 = (MN)^{-1} $   Nota: la potencia -1 indica inversa.

```python
# Escriba aquí su código
Q1=np.linalg.inv(M@N)
Q1
```

Calcule:  
    
$ Q2=(M^T N^T)^{-1}    $  el superíndice T significa Transpuesta

```python
# Escriba aquí su código 
T_M=M.T
T_N=N.T
Q2=np.linalg.inv(T_M@T_N)
print (T_M)
print (T_N)
#Dice que es una matriz singular lo cual significa que su determinante es 0, su inversa no existe
```

## Ej 4:


Cree la matriz P de números aleatorios (entre 0 y 1) de 100 filas x 200 columnas. Use el valor 123 como semilla para generar los números aleatorios:

```python
np.random.seed([123])
P=np.random.random((100,200))
P
```

Guarde en la variable P_min el menor valor entre todos los elementos de la matriz P:

```python
# Escriba aquí su código

P_min=np.min(P)
P_min
```

Cuál es el mayor valor de **cada una de las columnas** de P? Guarde los valores en P_max

```python
# Escriba aquí su código
P_max=np.max(P,axis=0)
P_max
```

Guarde en la variable n1 la cantidad de elementos de P cuyo valor es menor o igual que 0.3

```python
# Escriba aquí su código
n1=np.sum(P<=0.3)
n1
```

Cuál es el valor promedio de todos los elementos de P? Guardelo en P_prom

```python
# Escriba aquí su código

P_prom=np.mean(P)
P_prom
```

Cuántos elementos de P son mayores que dicho promedio? Guarde el resultado en P_may

```python
# Escriba aquí su código
P_may=np.sum(P>P_prom)
P_may


```

En cuántas columnas de P su promedio está por debajo del promedio de toda la matriz? Guarde en valor en n_menor

```python
# Escriba aquí su código
prom_col=np.mean(P,axis=0)
n_menor=np.sum(prom_col<P_prom)
n_menor

```

Cuántos valores de P son mayores o iguales que 0.2 y menores o iguales que 0.8? Guarde el resultado en la variable n2

```python
# Escriba aquí su código

n2=np.sum(np.logical_and(P>=0.2,P<=0.8))
n2
```

Cree la matriz P1, que se obtiene reemplazando todos los valores P que son menores o iguales que 0.5 por el valor 0 y reemplazando todos los valores mayores que 0.5 por el valor 1

```python
# Escriba aquí su código
P1=np.logical_and(np.where(P<=0.5,P,0), np.where(P>=0.5,P,1))
P1

```
